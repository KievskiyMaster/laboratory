```python
import re

# Функція для перетворення віку у числовий формат з урахуванням місяців і років
def convert_age(age):
    age = age.strip()
    if 'Months' in age:
        # Перетворюємо вікові діапазони в місяцях на середнє значення
        match = re.search(r'(\d+)-(\d+)', age)
        if match:
            start, end = map(int, match.groups())
            return (start + end) / 2 / 12  # Конвертуємо в роки
    elif 'Year' in age:
        # Перетворюємо "01 Year" і "02 Years" у числове значення
        match = re.search(r'(\d+)', age)
        if match:
            return float(match.group(1))
    return np.nan

# Застосовуємо функцію до стовпця "Age Years"
df['Age Years'] = df['Age Years'].apply(convert_age)

# Перевіримо результат
df[['Age Years']].head(10)
```

```python
import matplotlib.pyplot as plt

# Функція для категоризації віку
def categorize_age(age):
    if age < 13:
        return 'Child'
    elif age < 18:
        return 'Teenager'
    elif age < 65:
        return 'Adult'
    else:
        return 'Elderly'

# Створюємо новий стовпець з віковими групами
df['Age Group'] = df['Age Years'].apply(categorize_age)

# Підрахуємо кількість смертей у кожній віковій групі
age_group_counts = df.groupby('Age Group')['Total deaths'].sum()

# Побудуємо гістограму
age_group_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Deaths by Age Group')
plt.ylabel('Number of Deaths')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.show()
```
